```python
import tensorflow as tf
```

```python
daily_sales_number = [21,22,-108,31,-1,32,34,51]
```

```python
tf_dataset = tf.data.Dataset.from_tensor_slices(daily_sales_number)
tf_dataset
```

```python
for sale  in tf_dataset:
  print(sale.numpy())
```

```python
for sale  in tf_dataset.as_numpy_iterator():
  print(sale)
```

```python
# check first 3 elements
for sale in tf_dataset.take(3):
  print(sale.numpy())
```

```python
# filter invalid data points

tf_dataset = tf_dataset.filter(lambda x: x>0)
for sale in tf_dataset.as_numpy_iterator():
  print(sale)
```

```python
# change dollars into rupees
tf_dataset = tf_dataset.map(lambda x: x*72)
for sale in tf_dataset.as_numpy_iterator():
  print(sale)
```

```python
# shuffle
tf_dataset = tf_dataset.shuffle(2)
for sale in tf_dataset.as_numpy_iterator():
  print(sale)
```

```python
#  batching the training samples
for batch in tf_dataset.batch(2):
  print(batch.numpy())
```

```python
# all these operations in one single line
tf_dataset = tf.data.Dataset.from_tensor_slices(daily_sales_number)
tf_dataset = tf_dataset.filter(lambda x: x>0).map(lambda y: y*72).shuffle(2).batch(2)
```

```python
for batch in tf_dataset.as_numpy_iterator():
  print(batch)
```
